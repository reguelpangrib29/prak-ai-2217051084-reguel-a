# Linear Classifier

## Studi Kasus

Membuat klasifikasi data kualitas Wine menggunakan model klasifikasi linear dengan akurasi minimal 60%

## Library

```python
import pandas as pd   # Mengimpor library pandas dengan alias pd untuk manipulasi data tabular.
import numpy as np    # Mengimpor library numpy dengan alias np untuk operasi numerik.
import sklearn        # Mengimpor library sklearn untuk machine learning.

# Mengimpor kelas LogisticReggresion dari modul linear_model di sklearn untuk membuat model regresi logistik.
from sklearn.linear_model import LogisticRegression

# Mengimpor fungsi train_test_split dari modul model_selection di sklearn untuk membagi data menjadi set latih dan set uji.
from sklearn.model_selection import train_test_split

# Mengimpor fungsi accuracy_score dari modul metrics di sklearn untuk mengukur akurasi model.
from sklearn.metrics import accuracy_score
```

## Dataset

```python
# Membaca data dari URL CSV ke dalam DataFrame menggunakan pandas.
data = pd.read_csv("https://raw.githubusercontent.com/adivii/datasets/main/Artificial%20Intelligence/Pertemuan%205/Wine_Quality_Data.csv")

data # Menampilkan DataFrame yang berisi data Wine Quality
```

## Preprocessing dan Splitting Data

```python
# Import SimpleImputer
from sklearn.impute import SimpleImputer

# Penanganan data yang hilang (mengisi nilai yang hilang dengan nilai rata - rata)
imputer = SimpleImputer(strategy='mean')
imputer_data = imputer.fit_transform(data)

# Kembalikan data ke DataFrame
final_data = pd.DataFrame(imputer_data, columns=data.columns)

# Tampilkan hasil preprocessing
final_data
```

```python
# final_data diatas sudah final, disini hanya memperlihatkan apa yang terjadi pada data setelah menggunakan scaling minmax

# import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# Normalisasi fitur
normalizer = MinMaxScaler()
normalized_data = normalizer.fit_transform(imputer_data)

# Kembalikan data ke DataFrame
norm_data = pd.DataFrame(normalized_data, columns=data.columns)

# Tampilkan hasil preprocessing
norm_data
```

```python
# Memisahkan kolom target (quality) dan kolom fitur lainnya
X = final_data.drop("quality", axis=1)
y = final_data["quality"]
```

```python
# Normalisasi kolom fitur menggunakan MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Kolom target (quality) tidak di-normalisasi
y_categorical = pd.Series(pd.cut(y, bins=len(y.unique()), labels=False))
```

```python
# Memisahkan data menjadi data latih dan data uji
# X_train: fitur - fitur untuk data latih
# X_test: fitur - fitur untuk data uji
# y_train: target untuk data latih
# y_test: target untuk data uji
# test_size: proporsi data uji terhadap total data, disini diatur sebesar 20% (0.2)
# random_state: seed untuk mengontrol randomization. memastikan hasil yang reprodusibel
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_categorical, test_size=0.2, random_state=42)
```

## Modelling and Prediction

```python
# Membuat model Logistic Regression
# max_iter: Jumlah iterasi maksimum yang digunakan oleh algoritma optimasi untuk menemukan parameter model.
# Nilai 7124 disini adalah contoh angka sembarang dan bukan angka mainstream.
# Anda dapat menyesuaikan nilai ini sesuai kebutuhan atau menggunakan nilai default jika tidak yakin.
model = LogisticRegression(max_iter=7124)

# Melatih model menggunakan data latih
model.fit(X_train, y_train)
```

```python
#prediksi data test
y_pred = model.predict(X_test)
```

```python
# menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi model logistic regression: ", accuracy*100, "%")
```

Dari perhitungan akurasi model logistic regression mencapai 62% atau sudah mencapai target studi kasus.
Namun kita tetap perlu evaluasi model jika perlu menaikkan akurasi prediksinya agar lebih tinggi.

## Comparing Models

```python
# Compare ke model Support Vector Machines
from sklearn.svm import SVC

# Deklarasi model SVM
svm_model = SVC()

# Melatih model SVM
svm_model.fit(X_train, y_train)

# Evaluasi akurasi model
svm_accuracy = svm_model.score(X_test, y_test)

print("Akurasi model SVM: ", svm_accuracy*100, "%")
```

Akurasi nya naik sedikit menjadi 62.9% menggunakan model Support Vector Machines

```python
# Compare ke model Naive Bayes
from sklearn.naive_bayes import GaussianNB

# Deklarasi model SVM
nb_model = GaussianNB()

# Melatih model SVM
nb_model.fit(X_train, y_train)

# Evaluasi akurasi model
nb_accuracy = nb_model.score(X_test, y_test)

print("Akurasi model Naive Bayes: ", nb_accuracy*100, "%")
```

Menggunakan model Naive Bayes, akurasinya sedikit turun menjadi 61.1%
